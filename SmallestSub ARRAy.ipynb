```python
from collections import defaultdict
import sys
S = input()
dc = len(set(list(S)))
n = len(S)
f = defaultdict(int)
si = 0
ml = sys.maxsize
dth = 0 
for j in range(n):
    f[S[j]] += 1
    if f[S[j]] == 1:
        dth += 1

    if dc== dth:
        while f[S[si]] > 1:
            if f[S[si]] > 1:
                f[S[si]] -= 1
                si += 1

        cl = j - si + 1
        ml = min(ml, cl)
out = ml
print (out)
```
